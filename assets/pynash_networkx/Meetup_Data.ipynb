## Get data from Meetup.com

We are going to get data from Meetup's API. You can create custom requests using their API console, then make them via Python by copying the secure link here.

https://secure.meetup.com/meetup_api/console/?path=/2/members

```python
import requests
import pandas as pd
import numpy as np

df = pd.DataFrame()
for page in np.arange(7):
    q = 'https://api.meetup.com/2/members?offset={}&format=json&offset=3&group_urlname=PyNash&photo-host=public&page=200&key=1eb16676d664fa48314391ae5b6c'.format(page)
    response_from_meetup = requests.get(q)
    tdf = pd.DataFrame.from_dict(response_from_meetup.json()['results']).set_index('id')
    
    df = df.append(tdf)
```
